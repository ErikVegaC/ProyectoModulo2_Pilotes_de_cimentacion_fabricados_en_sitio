### **Planteamiento del problema**

En la construcción de estructuras de concreto reforzado, los pilotes desempeñan un papel fundamental al transferir las cargas de la superestructura hacia capas profundas y estables del suelo. El diseño y fabricación de pilotes deben cumplir con especificaciones técnicas rigurosas para garantizar la estabilidad y seguridad de las estructuras. Sin embargo, en la práctica, la variabilidad en las dimensiones de los pilotes y la disposición del refuerzo puede llevar a errores que impacten negativamente en la calidad y resistencia estructural, incrementando los costos de fabricación y material.

La optimización de los recursos y la reducción de errores en la fabricación de pilotes requiere un enfoque que considere la variabilidad de los parámetros geométricos y de refuerzo, así como su impacto en los volúmenes de concreto y acero utilizados. Además, el diseño de los pilotes debe cumplir con los requisitos de normas estructurales en términos del porcentaje de acero requerido para asegurar la estabilidad y resistencia de la estructura.

  - Objetivos específicos:
    - Implementar una función de simulación de Monte Carlo que genere aleatoriamente los parámetros geométricos y de refuerzo de un pilote, incluyendo diámetros, longitudes, número de varillas y dimensiones de los estribos.

    - Calcular los volúmenes de concreto y de acero en cada simulación, considerando tanto las varillas longitudinales como los estribos.

### **2. Modelo que representa el problema**


En este proyecto, se utiliza la técnica de simulación de Monte Carlo para evaluar la variabilidad de los parámetros geométricos y estructurales de un pilote reforzado con acero. El modelo representa matemáticamente los volúmenes de concreto y acero mediante las siguientes ecuaciones:

- **Volumen del pilote (cilindro)**:

$$ V_{\\{pilote}} = \pi (\frac{D}{2})^2  L $$

Donde:
- \\( D \\) es el diámetro del pilote.
- \\( L \\) es la longitud del pilote.

- **Volumen de una varilla (cilindro)**:

$$ V_{{varilla}} = \pi  (\frac{D_{barra}}{2})^2 L $$

Donde:
- \\( D_{{barra}} \\) es el diámetro de una barra de refuerzo.

- **Volumen total de las varillas**:

$$ V_{{totalvarillas}} = n_{{barras}}  V_{{varilla}} $$

Donde:
- \\( n_{{barras}} \\) es el número total de barras de refuerzo.

- **Volumen de un estribo (cilindro delgado)**:

$$ V_{{estribo}} = \pi (\frac{D_{{estribo}}}{2})^2  P_{{estribo}} $$

Donde:
- \\( D_{{estribo}} \\) es el diámetro del estribo.
- \\( P_{{estribo}} \\) es el perímetro de un estribo, calculado como \\( P_{{estribo}} = \pi D_{{pilote}} \\).

- **Volumen total de los estribos**:

$$ V_{{totalestribos}} = n_{{estribos}}  V_{{estribo}} $$

Donde:
- \\( n_{{estribos}} \\) es el número total de estribos a lo largo del pilote, calculado como \\( n_{{estribos}} = \frac{L}{{distanciaestribos}} \\).

- **Porcentaje de acero en el pilote**:

$$ {Porcentaje \ de \ acero} = \frac{V_{{totalacero}}}{V_{{pilote}}} * 100 $$

Donde:
- \\( V_{{totalacero}} = V_{{totalvarillas}} + V_{{totalestribos}} \\) es el volumen total de acero.

#### **Deducción del modelo**
El modelo se basa en las ecuaciones geométricas básicas de cilindros para calcular los volúmenes de los diferentes elementos estructurales. Estas ecuaciones se utilizan de manera iterativa en un enfoque de Monte Carlo para incorporar variabilidad en los parámetros geométricos (diámetros, longitudes y número de barras y estribos).

El método de Monte Carlo se implementa generando valores aleatorios para cada parámetro de interés dentro de un rango específico, considerando la incertidumbre en el diseño estructural. Cada iteración representa una configuración hipotética de un pilote, y se calculan los volúmenes y porcentajes de acero correspondientes.

#### **Situación que representa el modelo**
Este modelo representa la situación de diseño de un pilote reforzado con acero, donde se desea evaluar cómo la variabilidad de las dimensiones del pilote y su refuerzo afecta los volúmenes de concreto y acero. La finalidad es obtener una distribución estadística de los volúmenes y un porcentaje promedio de acero, para asegurar que el diseño cumpla con los estándares de ingeniería y normativas.

#### **Limitaciones Fundamentales del Modelo**
- **Simplificación Geométrica**: Se asume que tanto el pilote como las varillas y los estribos tienen formas cilíndricas perfectas. No se consideran irregularidades o deformaciones.


- **Distribuciones Uniformes**: Se utilizan distribuciones uniformes para los parámetros, lo cual puede no reflejar adecuadamente la variabilidad real de los materiales y procesos de construcción.


- **Incertidumbre del Número de Estribos**: El modelo asume que la distancia entre estribos es constante, lo cual puede no ser realista en la construcción.

#### **Significado y valor de los parámetros**
- **Diámetro del Pilote (\\( D \\))**: Representa el diámetro del pilote de concreto. Su rango se define según especificaciones estructurales (0.94 a 1 m en este caso).


- **Longitud del Pilote (\\( L \\))**: Define la longitud del pilote, usualmente determinada por las condiciones del terreno (10 a 10.5 m en este caso).


- **Número de Barras de Refuerzo (\\( n_{{barras}} \\))**: Número total de barras longitudinales de refuerzo.


- **Diámetro de las Barras (\\( D_{{barra}} \\))**: Diámetro de las barras longitudinales de refuerzo.


- **Diámetro del Estribo (\\( D_{{estribo}} \\))**: Diámetro de cada estribo que rodea las barras longitudinales.


- **Distancia entre Estribos**: Distancia entre dos estribos consecutivos a lo largo de la longitud del pilote.
"""


```python
import numpy as np
import matplotlib.pyplot as plt


def calcular_volumen_montecarlo(n_simulaciones):
    resultados_concreto = []
    resultados_acero = []
    resultados_porcentaje = []

    for i in range(n_simulaciones):

        D = np.random.uniform(0.94, 1)
        L = np.random.uniform(10, 10.5)
        num_barras = np.random.randint(7,8)
        diam_barra = np.random.uniform(0.02, 0.022)
        diam_estribo = np.random.uniform(0.008, 0.012)
        distancia_estribos = np.random.uniform(0.3, 0.35)
# V pilote
        radio_pilote = D / 2
        volumen_concreto = np.pi * radio_pilote**2 * L

# V varillas
        radio_barra = diam_barra / 2
        volumen_barra = np.pi * radio_barra**2 * L
        volumen_total_acero = num_barras * volumen_barra

# C estribos
        num_estribos = int(L / distancia_estribos)

# V estribos
        radio_estribo = diam_estribo / 2
        longitud_estribo = np.pi * D
        volumen_estribo = np.pi * radio_estribo**2 * longitud_estribo
        volumen_total_estribos = num_estribos * volumen_estribo

# % acero
        volumen_total_acero += volumen_total_estribos
        porcentaje_acero = (volumen_total_acero / volumen_concreto) * 100

        resultados_concreto.append(volumen_concreto)
        resultados_acero.append(volumen_total_acero)
        resultados_porcentaje.append(porcentaje_acero)

    return resultados_concreto, resultados_acero, resultados_porcentaje


n_simulaciones = 10000


resultados_concreto, resultados_acero, resultados_porcentaje = calcular_volumen_montecarlo(n_simulaciones)


promedios_concreto = np.cumsum(resultados_concreto) / np.arange(1, n_simulaciones + 1)
promedios_acero = np.cumsum(resultados_acero) / np.arange(1, n_simulaciones + 1)
promedios_porcentaje = np.cumsum(resultados_porcentaje) / np.arange(1, n_simulaciones + 1)


```

### **3 - Solución del problema de optimización**

En el presente trabajo se ha abordado el problema de estimar los volúmenes de concreto y acero en pilotes reforzados utilizando el método de Monte Carlo. La técnica de Monte Carlo permite simular múltiples configuraciones estructurales mediante la variación de parámetros clave, como los diámetros del pilote y del refuerzo, la longitud del pilote, el número de barras y la distancia entre estribos.
Solución del problema utilizando el método de Monte Carlo

El modelo desarrollado genera configuraciones aleatorias para los parámetros geométricos del pilote, respetando los rangos definidos según especificaciones estructurales. Posteriormente, en cada simulación se calculan los volúmenes de concreto y acero, y el porcentaje de acero respecto al volumen total. El algoritmo utiliza las siguientes funciones:

  - Generación aleatoria de parámetros: Se generan valores aleatorios dentro de un rango establecido para cada dimensión y cantidad de refuerzo.


  - Cálculo de volúmenes: Se calculan los volúmenes de concreto, barras longitudinales y estribos usando ecuaciones de geometría cilíndrica.


  - Evaluación del porcentaje de acero: Se calcula el porcentaje de acero en cada iteración para evaluar si cumple con los límites deseados o especificaciones de diseño.


  - Análisis de convergencia: Los promedios acumulados de las simulaciones se utilizan para verificar si los resultados se estabilizan y conducen a una solución confiable.

#### **¿Condujo el algoritmo a una solución factible?**

Los resultados obtenidos del método de Monte Carlo se analizaron mediante gráficos de convergencia, que mostraron cómo los volúmenes de concreto, el volumen total de acero y el porcentaje de acero se estabilizan con un número suficiente de simulaciones. Esto indica que el algoritmo de Monte Carlo ha logrado proporcionar una solución factible al problema planteado, ya que los resultados promedio cumplen con los estándares de diseño y normativas estructurales.

Los volúmenes calculados para concreto y acero, así como el porcentaje de acero, se encuentran dentro de los límites esperados, lo que valida la efectividad del algoritmo. Además, la flexibilidad del método permite ajustar los rangos de los parámetros para optimizar el diseño de los pilotes en función de distintas especificaciones o condiciones del terreno.

 ### **4 - Visualización de la solución del problema**

```python

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(promedios_concreto, color='b')
plt.title("Volumen de Concreto")
plt.xlabel("Simulaciones")
plt.ylabel("Volumen Promedio de Concreto (m3)")
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(promedios_acero, color='r')
plt.title("Volumen de Acero")
plt.xlabel("Simulaciones")
plt.ylabel("Volumen Promedio de Acero (m3)")
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(promedios_porcentaje, color='g')
plt.title("Porcentaje de Acero")
plt.xlabel("Simulaciones")
plt.ylabel("Porcentaje Promedio de Acero (%)")
plt.grid()

plt.tight_layout()
plt.show()

```
